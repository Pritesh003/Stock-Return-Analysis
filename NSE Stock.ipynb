### Importing necessary libraries

```python
from nsepy import get_history
from datetime import date
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### Getting data of two of the biggest Pharma companies from nsepy.

```python
SunPharma = get_history(symbol='SUNPHARMA', start=date(2019,3,1), end=date(2020,2,29))
Lupin = get_history(symbol='LUPIN', start=date(2019,3,1), end=date(2020,2,29))
```

### Saving the data in .csv file for future use.

```python
SunPharma.to_csv('SunPharma.csv')
Lupin.to_csv('Lupin.csv')
```

#### Taking a look at the data sets

```python
SunPharma
```

### Calculating return for both the data sets using the closing price of stock.

```python
SunPharma['Return'] = SunPharma['Close'] - SunPharma['Prev Close']
SunPharma['%Return'] = SunPharma['Close'].pct_change()
SunPharma.head()
```

```python
Lupin['Return'] = Lupin['Close'] - Lupin['Prev Close']
Lupin['%Return'] = Lupin['Close'].pct_change()
Lupin.head()
```

### Plotting the stock price and return for both the stocks.

```python
f, axes = plt.subplots(2,1, figsize=(18,8))
axes[0].plot(SunPharma['Close'], color='orange', marker='o', markersize=3, linewidth=1)
axes[1].plot(Lupin['Close'], color='blue', marker='o', markersize=3, linewidth=1)
axes[0].legend(['Sun Pharma'])
axes[1].legend(['Lupin'])
plt.suptitle('Closing Price', size=24)
plt.show()
```

```python
f, axes = plt.subplots(2,1, figsize=(18,8))
axes[0].plot(SunPharma['Return'], color='orange', marker='o', markersize=3, linewidth=1)
axes[1].plot(Lupin['Return'], c='blue', marker='o', markersize=3, linewidth=1)
axes[0].legend(['Sun Pharma'])
axes[1].legend(['Lupin'])
plt.suptitle('Return', size=24)
plt.show()
```

```python
f, axes = plt.subplots(1,2, figsize=(18,6))
axes[0].hist(SunPharma['Return'], color='orange')
axes[1].hist(Lupin['Return'], color='blue')
axes[0].legend(['Sun Pharma'])
axes[1].legend(['Lupin'])
plt.suptitle('Return', size=24)
plt.show()
```

### Question: Why stock return distribution is not a Normal curve?
The stock return plot above looks normal but if we look to the left of the distribution, we can see fat tails. The fat tails mean that extreme events occur more frequently in reality than what a normal distribution would predict. The actual distribution of the returns is skinnier in its center than the normal distribution. The skinny middle and the fat tails imply that the normal distribution might not be the best describer of stock returns. 

#### Mean

```python
SP_mean = SunPharma['Return'].mean()
LP_mean = Lupin['Return'].mean()

print('Mean return for Sun Pharma is {:.4f} for Mar-2019 to Feb-2020'.format(SP_mean))
print('Mean return for Lupin is {:.4f} for Mar-2019 to Feb-2020'.format(LP_mean))
```

#### Variance

```python
SP_var =  SunPharma['Return'].var()
LP_var = Lupin['Return'].var()

print('Variance for Sun Pharma is {:.4f} for Mar-2019 to Feb-2020'.format(SP_var))
print('Variance for Lupin is {:.4f} for Mar-2019 to Feb-2020'.format(LP_var))
```

### Question: Why we use variance as a risk measure?
Variance measures variability from the average or mean. To investors, variability is volatility, and volatility is a measure of risk. Variance is of paramount importance as they are used to measure security and market volatility, which in turn plays a large role in creating a profitable trading strategy.

#### Skewness

```python
SP_skew =  SunPharma['Return'].skew()
LP_skew = Lupin['Return'].skew()

print('Skewness for Sun Pharma is {:.4f} for Mar-2019 to Feb-2020'.format(SP_skew))
print('Skewness for Lupin is {:.4f} for Mar-2019 to Feb-2020'.format(LP_skew))
```

#### Kurtosis

```python
SP_kurt =  SunPharma['Return'].kurtosis()
LP_kurt = Lupin['Return'].kurtosis()

print('Kurtosis for Sun Pharma is {:.4f} for Mar-2019 to Feb-2020'.format(SP_kurt))
print('Kurtosis for Lupin is {:.4f} for Mar-2019 to Feb-2020'.format(LP_kurt))
```

### Correlation between the stocks.

```python
data = {'SP_Close': SunPharma['Close'], 'Lup_Close': Lupin['Close']}
data = pd.DataFrame(data)
data
```

```python
data.corr()
```

```python
plt.figure(figsize=(10,6))
sns.regplot('SP_Close', 'Lup_Close', data = data, marker = '+')
plt.title('Regression Plot', size=18)
plt.show()
```

### Question: Causation vs Correlation.
Correlation is a statistical measure (expressed as a number) that describes the size and direction of a relationship between two or more variables. A correlation between variables, however, does not automatically mean that the change in one variable is the cause of the change in the values of the other variable.

Causation indicates that one event is the result of the occurrence of the other event; i.e. there is a causal relationship between the two events. This is also referred to as cause and effect.

#### Correlation doesn't imply causation.
We can see that there is a positive correlation between both the stocks. Both stocks decrease and increase somewhat togather, but it doesn't imply that decrease in return of any one stock will cause the return of other stock to drop. There could be a third factor that influence the stock price for both the stocks in a similar way so that it appears to decrease or increase together.
